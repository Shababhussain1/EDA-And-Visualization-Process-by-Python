# **Exploratory Data Analysis and Visualization by Using Pandas and Seaborn**

```python
#Importing the Required Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv("/content/titanic_dataset.csv")
```

```python
df.dtypes
```

```python
df.shape
```

```python
df.count()
```

```python
df = df.drop_duplicates()
df.head(5)
```

```python
print(df.isnull().sum())
```

```python
df = df.dropna()    # Dropping the missing values.
df.count()
```

```python
print(df.isnull().sum())   # After dropping the values
```

```python
df.head()
```

```python
df.tail()
```

# **Distplot**

```python
sns.histplot(df['Parch'],kde=False)
plt.show()
```

# **KdePlot**

```python
sns.distplot(df['Age'], hist=False)
plt.show()
```

```python
plt.figure(figsize=(8,8))
sns.distplot(df['Age'])
plt.show()
```

# **Scatterplot**

```python
plt.figure(figsize=(8,8))
sns.scatterplot(x="Age", y="Fare", hue="Sex", data=df)
plt.show()
```

# **Relplot**

```python
sns.relplot(x="Age", y="Fare", col="Pclass", hue="Sex", style="Sex",kind="line", data=df) # scatter can be used instead of "line" plot
plt.show()
```

# **Lineplot**

```python
plt.figure(figsize=(8,8))
sns.lineplot(x="Age", y="Fare", hue="Sex", style="Sex", data=df)
plt.show()
```

# **Barplot**

```python
plt.figure(figsize=(8,8))
sns.barplot(x="Sex", y="Survived", hue="Pclass", data=df)
plt.show()
```

# **Stripplot**

```python
plt.figure(figsize=(8,8))
sns.stripplot(x="Sex", y="Age",hue='Sex', data=df)
plt.show()
```

# **SwarmPlot**

```python
plt.figure(figsize=(8,8))
sns.swarmplot(x="Sex", y="Age",hue='Sex', data=df)
plt.show()
```

# **Boxplot**

```python
plt.figure(figsize=(8,8))
sns.boxplot(x="Survived", y="Age", data=df)
plt.show()
```

# **Violinplot**

```python
sns.violinplot(x="Survived", y="Age", hue='Sex', data=df)
plt.show()
```

# **Countplot**

```python
sns.countplot(x="Survived", data=df, palette="Blues");
plt.show()
```

# **Pointplot**

```python
plt.subplots(figsize=(8, 8))
sns.pointplot(x="Sex", y="Survived", hue="Pclass", data=df)
plt.show()
```

# **Heatmap**

```python
# Drop unwanted columns
df_cleaned = df.drop(columns=["Name", "Sex", "Ticket", "Cabin", "Embarked"])

# Plot the heatmap for the cleaned DataFrame
plt.subplots(figsize=(10, 10))
sns.heatmap(df_cleaned.corr(), cmap="YlGnBu", annot=True, fmt=".2f")
plt.show()
```
